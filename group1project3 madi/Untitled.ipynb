```python
import requests
from bs4 import BeautifulSoup
```

```python
# Send a GET request to the website
url = "https://www.worldometers.info/coronavirus/"
response = requests.get(url)
```

```python
# Create a BeautifulSoup object to parse the HTML content
soup = BeautifulSoup(response.text, "html.parser")
```

```python


# Find the table containing the country-wise data
table = soup.find("table", id="main_table_countries_today")

# Find all rows in the table
rows = table.find_all("tr")

# Iterate over the rows and extract the data
for row in rows:
    # Find all columns in the row
    columns = row.find_all("td")
    
    # Extract the desired information (cases, deaths, population, etc.) from the columns
    if len(columns) >= 4:
        country = columns[0].text.strip()
        cases = columns[1].text.strip()
        deaths = columns[3].text.strip()
        population = columns[13].text.strip()
        
        # Process the extracted data as needed (e.g., store in a database, perform calculations, etc.)
        print("Country:", country)
        print("Cases:", cases)
        print("Deaths:", deaths)
        print("Population:", population)
        print("--------------------")
```
